**Step 1: Importing the libraries**

```python
import pandas as pd
import numpy as np
```

**Step 2: Importing dataset**

```python
df=pd.read_csv('Data.csv')
df.head()
```

**Step 3: Handling the missing data**

```python
df.isnull().sum()
```

```python
# filled the missing value by taking mean based on the country
sum1=df.loc[df['Country'] == 'Spain', 'Age'].sum()
sum1//=3
a=df.loc[df['Country'] == 'Germany', 'Salary'].sum()
a//=3
df['Age']=df['Age'].fillna(sum1)
df['Salary']=df['Salary'].fillna(a)
df
```

**Step 4: Encoding categorical data**

```python
# implemented label encoder to convert purchase column 
# Import label encoder
from sklearn import preprocessing

# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()

# Encode labels in column 'species'.
df['Purchased']= label_encoder.fit_transform(df['Purchased'])

df

```

```python
# implemented one hot encoding to convert country column 
one_hot_encoded_data = pd.get_dummies(df, columns = ['Country'])
one_hot_encoded_data
```

**Step 5: Creating a dummy variable**

```python
dummies = pd.get_dummies(df, columns = ['Country'])
dummies
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test  = train_test_split(df.drop('Purchased',axis=1),df['Purchased'],test_size=0.2,random_state=45)
print('x_train',X_train.shape,'x_test',X_test.shape,'y_train',y_train.shape,'y_test',y_test.shape)
```

**Step 7: Feature Scaling**

```python
# importing the required libraries
from sklearn.preprocessing import MinMaxScaler

# Creating an instance of the sklearn.preprocessing.MinMaxScaler()
# specifying the min and max value of the scale
scaler = MinMaxScaler(feature_range=(10, 100,10))

# Scaling the Price column of the created dataFrame
# and storing the result in ScaledPrice Column
df[["Salary"]] = scaler.fit_transform(df[["Salary"]])

print(df)
```
